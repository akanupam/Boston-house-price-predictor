```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/BostonHousing.csv"
```

```python
boston = pd.read_csv(url)
```

```python
boston.keys()
```

```python
dataset = pd.DataFrame(boston)
```

```python
dataset.head()
```

```python
dataset = dataset.rename(columns={'medv': 'price'})
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
# Summarizing stats of data
dataset.describe()
```

```python
# check for missing values
dataset.isnull().sum()
```

```python
# exploratory data analysis
# correlation
dataset.corr()
```

```python
import seaborn as sns
sns.pairplot(dataset)
```

```python
plt.scatter(dataset.crim, dataset.price)
plt.xlabel('crime rate')
plt.ylabel('price')
plt.show()
```

```python
plt.scatter(dataset.rm, dataset.price)
plt.xlabel('rm')
plt.ylabel('price')
plt.show()
```

```python
import seaborn as sns
sns.regplot(x=dataset.rm, y=dataset.price)
```

```python
sns.regplot(x=dataset.lstat, y=dataset.price)
```

```python
#dividing independent and dependent features
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)
```

```python
X_train.head()
```

```python
#standardizing dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test) #we are not using fir becoz we dont want our test data to get affected by training dataset
X_train
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# on which parameters the model has been prepared
reg.get_params()
```

```python
# prediction with test data
reg_pred = reg.predict(X_test)
reg_pred
```

```python
#plot a scatter plot for prediction
plt.scatter(reg_pred, y_test)
```

```python
# prediction with residuals
residuals = y_test - reg_pred
residuals
```

```python
#plotting residuals
sns.displot(residuals, kind='kde')
```

```python
#plot wrt prediction and residuals
#uniform distributios
plt.scatter(reg_pred, residuals)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```
